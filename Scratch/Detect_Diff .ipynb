# Filter Data

```python
#Function filter data per day
def excludeDataDay(data,c):
    data2 = data[(data['NW (Kg)'] > 0.0) & (data['Value (IDR)'] != 0) ]  #Exclude Free Goods
    data3 = data2[data2['Customer'] != 2064741]     #Exclude Paseo Lady
    data4 = data3[data3['GI Date'] == c]
    print("Data Dimension ", data4.shape)
    print("Sum Value ", sum(data4['Value (IDR)']))
    print("Sum Weight ", sum(data4['NW (Kg)']))
    return data4
```

```python
#Function filter Interval data
def excludeIntervalData(data,c):
    data2 = data[(data['NW (Kg)'] > 0) & (data['Value (IDR)'] != 0) ]  #Exclude Free Goods
    data3 = data2[data2['Customer'] != 2064741]     #Exclude Paseo Lady
    data4 = data3[c]
    print("Data Dimension ", data4.shape)
    print("Sum Value ", sum(data4['Value (IDR)']))
    print("Sum Weight ", sum(data4['NW (Kg)']))
    return data4
```

# Find Diff 

### One Year Diff

```python
#2018, 01 - 07 Desember Data
c = (dfOur2['GI Date']>='2018-01-01 ') & (dfOur2['GI Date'] <= '2018-12-31')
d = (dfUser3['GI Date']>='2018-01-01') & (dfUser3['GI Date'] <= '2018-12-31')

dfOur18 = excludeIntervalData(dfOur2,c)
print()
print("dfUser 2018 :")
dfUser18 = excludeIntervalData(dfUser3,d)
print()
print(sum(dfOur18['Value (IDR)']) - sum(dfUser18['Value (IDR)']))
sum(dfOur18['NW (Kg)']) - sum(dfUser18['NW (Kg)'])
```

```python
a = dfOur18[["Material","SO No","Value (IDR)","GI Date", "NW (Kg)"]]
b = dfUser18[["Material","SO No","Value (IDR)","GI Date","NW (Kg)"]]
a = pd.pivot_table(a, values=['Value (IDR)','NW (Kg)'], index=['Material','SO No','GI Date'], aggfunc=np.sum).reset_index()
b = pd.pivot_table(b, values=['Value (IDR)','NW (Kg)'], index=['Material','SO No','GI Date'], aggfunc=np.sum).reset_index()
#a['Value (IDR)'].aggregate(np.sum)
#b['Value (IDR)'].aggregate(np.sum)
c = pd.merge(a, b, on=['Material','SO No'], how='inner')
c['Diff_Value'] = c['Value (IDR)_y'] - c['Value (IDR)_x']
c['Diff_Weight(Kg)'] = c['NW (Kg)_y'] - c['NW (Kg)_x']
```
